```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Scatter Plot (sns.scatterplot):
Use: Used to visualize the relationship between two continuous variables. It helps in identifying the correlation, trends, and outliers in the data.
Line Plot (sns.lineplot):
Use: Used to display data points connected by straight lines. It is commonly used to visualize trends over time.
Box Plot (sns.boxplot):
Use: Used to display the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, and maximum). It helps in identifying outliers and understanding the spread and skewness of the data.
Histogram (sns.histplot):
Use: Used to visualize the distribution of a single continuous variable. It shows the frequency of data points within specified bins, helping to understand the underlying distribution.
Heatmap (sns.heatmap):
Use: Used to visualize data in a matrix format with different colors representing different values. 
It is useful for displaying correlations, missing data, or any matrix-like data.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Box plot for age
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Histogram for price with hue for cut
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title("Price Distribution by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
# Load the iris dataset
iris = sns.load_dataset("iris")

# Pair plot
sns.pairplot(data=iris, hue='species')
plt.title("Pair Plot of Iris Dataset")
plt.show()

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the data for heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Flights Data")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
