```python
Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its
application.

Min-Max Scaling:

Definition: Min-Max scaling (or normalization) transforms the data to a fixed range, usually [0, 1] or [-1, 1].
Q2. What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling?
Provide an example to illustrate its application.
Unit Vector Technique:

Definition: The unit vector technique, also known as vector normalization, scales each feature 
vector to have a unit norm (i.e., a length of 1).
Difference from Min-Max Scaling: While Min-Max scaling scales features to a fixed range, the unit vector technique scales
features to have unit norm, preserving the direction of data points but not their magnitude.
Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an
example to illustrate its application.
Principal Component Analysis (PCA):

Definition: PCA is a technique used to reduce the dimensionality of a dataset while preserving as much variability as possible.
How it works:
Covariance Matrix: Compute the covariance matrix of the data.
Eigenvectors and Eigenvalues: Compute the eigenvectors and eigenvalues of the covariance matrix.
Principal Components: Select the top k eigenvectors corresponding to the largest eigenvalues to form the principal components.
Usage: PCA transforms the data into a new coordinate system where the axes (principal components) represent directions of maximum variance.
Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature
Extraction? Provide an example to illustrate this concept.
Relationship between PCA and Feature Extraction:

Feature Extraction: PCA can be used to create new features (principal components) that capture the most important information in the data.
PCA for Feature Extraction: By projecting the data onto the principal components, PCA can reduce
the dimensionality while retaining the most significant features.
Q5. You are working on a project to build a recommendation system for a food delivery service. The dataset
contains features such as price, rating, and delivery time. Explain how you would use Min-Max scaling to
preprocess the data.
Steps to Use Min-Max Scaling:

Load the Dataset: Ensure data is clean and preprocessed.
Apply Min-Max Scaling: Scale each feature to a fixed range, typically [0, 1] or [-1, 1].
Q6. You are working on a project to build a model to predict stock prices. The dataset contains many
features, such as company financial data and market trends. Explain how you would use PCA to reduce the
dimensionality of the dataset.
Steps to Use PCA:

Load the Dataset: Ensure data is clean and preprocessed.
Standardize the Data: Standardize features to have a mean of 0 and a variance of 1.
Apply PCA: Determine the number of principal components to retain based on the explained variance ratio.
Transform Data: Reduce the dimensionality of the dataset
Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the
values to a range of -1 to 1.
Steps:

Calculate Min and Max: Determine the minimum and maximum values in the dataset.
Apply Min-Max Scaling Formula: Transform the values to the desired range.
Q8. For a dataset containing the following features: [height, weight, age, gender, blood pressure], perform
Feature Extraction using PCA. How many principal components would you choose to retain, and why?
Steps:

Load the Dataset: Ensure data is clean and preprocessed.
Encode Categorical Variables: Convert categorical variables like gender to numerical values.
Standardize the Data: Standardize features to have a mean of 0 and a variance of 1.
Apply PCA: Determine the number of principal components to retain based on the explained variance ratio.
```
