Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask Framework:
Flask is a micro web framework for Python based on Werkzeug and Jinja2. It is designed to be simple, flexible, and lightweight, providing the necessary tools and libraries to build web applications quickly and efficiently. Flask is classified as a microframework because it does not require particular tools or libraries, making it easy to start and expand as needed.

Advantages of Flask Framework:

Simplicity and Flexibility:

Flask's minimalistic approach allows developers to start small and grow their applications as needed. It provides a simple core with extensibility through various plugins and extensions.
Lightweight:

As a microframework, Flask has a small core, which means it imposes little overhead. This makes it ideal for small projects or prototypes that do not require extensive features.

Q3. What is App routing in Flask? Why do we use app routes?
App Routing in Flask:
App routing in Flask is the mechanism used to map URLs to the functions within a web application. Each route is associated with a specific URL and when that URL is accessed, the corresponding function (also known as a view function or view) is executed. In Flask, routes are defined using the @app.route decorator.
We use App routing for below following reasons

URL Mapping:

App routes provide a way to define how URLs should be mapped to specific functions in the application. This makes it easy to organize and manage the different parts of the web application.
Separation of Concerns:

By mapping URLs to specific functions, Flask allows for a clear separation of concerns. Each function can handle a specific request and generate an appropriate response, making the codebase more modular and maintainable.
RESTful API Design:

App routes are essential for creating RESTful APIs. By defining routes for different HTTP methods (GET, POST, PUT, DELETE), Flask allows developers to build APIs that follow REST principles.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
In Flask, the url_for function is used for URL building. This function generates a URL to a specific function, ensuring that the URLs are built consistently and correctly throughout the application. The url_for function takes the endpoint name (typically the name of the view function) as its first argument and can also accept keyword arguments corresponding to the variables in the route.

Example of Using url_for
Here is an example demonstrating how to use url_for for URL building:
from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, World!'

@app.route('/about')
def about():
    return 'About Page'

@app.route('/user/<username>')
def profile(username):
    return f'User: {username}'

@app.route('/redirect-to-about')
def redirect_to_about():
    return redirect(url_for('about'))

@app.route('/user-profile/<username>')
def user_profile(username):
    # Use url_for to build the URL to the user's profile page
    profile_url = url_for('profile', username=username)
    return f'Profile URL for {username}: {profile_url}'

if __name__ == '__main__':
    app.run(debug=True)
