1) What is multithreading in python? Why is it used? Name the module used to handle threads in python

Multithreading in Python refers to the concurrent execution of two or more threads within a single process. A thread is a separate flow of execution, meaning that the program can perform multiple tasks simultaneously.

Why is Multithreading Used?

Parallelism: To perform multiple operations concurrently, which can improve the efficiency of a program.
Responsiveness: To keep applications responsive, particularly in GUI applications where long-running tasks can be executed in the background.
Resource Sharing: Threads share the same memory space and resources of the process, making it easier to share data between threads.
Module Used: The threading module is used to handle threads in Python.

4. Write a python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes

import threading

def print_squares():
    squares = [i**2 for i in range(10)]
    print("Squares:", squares)

def print_cubes():
    cubes = [i**3 for i in range(10)]
    print("Cubes:", cubes)


thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)


thread1.start()
thread2.start()


thread1.join()
thread2.join()


5. State advantages and disadvantages of multithreading

Advantages:

Improved Performance: Allows multiple tasks to be performed concurrently, potentially improving the performance of I/O-bound operations.
Responsiveness: In interactive applications, background threads can keep the application responsive.
Resource Sharing: Threads share the same memory space, making it easier to share data between threads compared to processes.
Disadvantages:

Complexity: Adds complexity to the code, making it harder to debug and maintain.
Thread Safety: Issues like race conditions can occur if threads access shared resources simultaneously.
GIL Limitation: In CPython, the Global Interpreter Lock (GIL) can limit the performance benefits of multithreading for CPU-bound tasks.
 
6. Explain deadlocks and race conditions.
Deadlocks:

Definition: A deadlock occurs when two or more threads are blocked forever, waiting for each other to release resources. This results in a situation where none of the threads can proceed.
Example: If Thread A holds Resource 1 and waits for Resource 2, while Thread B holds Resource 2 and waits for Resource 1, neither thread can proceed, resulting in a deadlock.
Race Conditions:

Definition: A race condition occurs when two or more threads access shared resources simultaneously, and the outcome depends on the order in which the threads are scheduled. This can lead to inconsistent or unexpected results.
Example: If two threads are incrementing a shared counter without proper synchronization, they might overwrite each other's changes, leading to incorrect results.
Avoiding Deadlocks and Race Conditions:

Deadlocks: Avoid holding multiple locks at the same time, use timeout mechanisms for acquiring locks, or implement a lock ordering protocol.
Race Conditions: Use synchronization mechanisms such as locks (threading.Lock), semaphores, or other concurrency control techniques to ensure that only one thread accesses shared resources at a time.
